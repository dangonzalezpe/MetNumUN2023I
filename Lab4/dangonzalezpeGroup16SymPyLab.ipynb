$$P_n(x) = 0 $$

### **Numerical complex evaluation Wolfram**

a = 2 * a1 + I * a2, b = 3 * b1 + I * b2 , c = 5 * c1 + I * c2 , a * b * c

image.png
image.png

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)  
x = x1 + I*x2
y = y1 + I*y2
z = z1 + I*z2

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```

### **Numerical complex evaluation SymPy**

```
a1, a2, b1, b2, c1, c2 = symbols("a1 a2 b1 b2 c1 c2", real=True)  
a = 2*a1 + I*a2
b = 3*b1 + I*b2
c = 5*c1 + I*c2

print(a*b*c)
print(expand(a*b*c))
print(expand((a*b)*c))
print(expand(a*(b*c)))
q = 1/(pi + I)
print('q=',q)
q=N(q)
print('q=',q)
```
