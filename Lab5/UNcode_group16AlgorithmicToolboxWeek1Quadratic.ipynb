```python
import math 
def solve_quad(b, c):
  """Solve a quadratic equation, x**2 + bx + c = 0.
    
    Parameters
    ----------
    b, c : float
       Coefficients
       
    Returns
    -------
    x1, x2 : float or complex roots.  """
    
  # ... enter your code here ...

  disc = b**2-(4*c)

  if(disc  == b**2):
      x1 = -b
      x2 = c/x1
  elif (disc >= 0):
      rdisc = math.sqrt(disc)
      x1 = (-b + rdisc)/2
      x2 = (-b - rdisc)/2
  else:
      rdisc = math.sqrt(-disc)
      x1 = complex(-b/2, rdisc/2)
      x2 = complex(-b/2, -rdisc/2)
       
  return x1,x2
```
